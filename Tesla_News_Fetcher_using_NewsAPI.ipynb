```python
import pandas as pd
import requests

response = requests.get('https://newsapi.org/v2/everything?q=tesla&from=2024-06-25&sortBy=publishedAt&apiKey=4f7666ea33c14fbdbadfd99b3095027b') # assign result of requests.get to response variable

data = response.json() # assign result of response.json() to data variable
# Check if the response contains an error message
if 'status' in data and data['status'] == 'error':
    print(f"Error: {data['code']} - {data['message']}")
else:
    # Proceed with accessing the results
    results = data['articles']
    print(results)
```

```python
pd.DataFrame(response.json()['articles']).head(2)
```

```python
df= pd.DataFrame(response.json()['articles'])[['description','url','content']]
```

```python
df.head(12)
```

```python
for i in range(1,200):

    response = requests.get('https://newsapi.org/v2/everything?q=tesla&from=2024-06-25&sortBy=publishedAt&apiKey=4f7666ea33c14fbdbadfd99b3095027b'.format(i))
    temp_df = pd.DataFrame(response.json()['articles'])[['description','url','content']]
    df = pd.concat([df, temp_df], ignore_index=True) # Use pd.concat instead of append
```

```python
# Remove the extra dot in the to_csv method
df.to_csv('news_Demo.csv')
```
